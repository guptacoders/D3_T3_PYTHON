# WEB SCRAPPING USING BEAUTIFUL SOUP

Subslikescript website

```python
import requests
from bs4 import BeautifulSoup
website="https://subslikescript.com/movie/Titanic-120338"
content=requests.get(website)
content
```

```python
soup=BeautifulSoup(content.text,"html.parser")
soup
```

```python
soup.find("h1").text.strip()
```

```python
script=soup.find("div",class_="full-script")
script=script.text.strip()
# print(script)
```

```python
with open("titanic.txt","w",encoding="utf-8") as f:
    f.write(script)
```

PolitiFact Website

```python
import requests
from bs4 import BeautifulSoup
website="https://www.politifact.com/factchecks/list/"
content=requests.get(website)
content
```

```python
soup=BeautifulSoup(content.text,"html.parser")
soup
```

```python
frame=[]
links=soup.find_all("li",class_="o-listicle__item")

for i in links:
    statement=i.find("div",class_="m-statement__quote").text.strip()
    link="https://www.politifact.com"
    link+=i.find("div",class_="m-statement__quote").find("a")["href"].strip()
    date=i.find("div",class_="m-statement__body").find("footer").text.strip()
    date=date.split("•")[-1]
    source=i.find('a',class_="m-statement__name").text.strip()
    frame.append((statement,link,date,source))
```

```python
import pandas as pd
df=pd.DataFrame(frame,columns=["statement","link","date","source"])
df
```

Indian Express

```python
import requests
from bs4 import BeautifulSoup
website="https://indianexpress.com/latest-news/?ref=latestnews_hp"
content=requests.get(website)
content
```

```python
soup=BeautifulSoup(content.text,"html.parser")
```

```python
headlines = soup.find_all("div", class_="title")

# Print each headline
for headline in headlines:
    title_tag = headline.find("a")
    if title_tag:
        print(title_tag.text.strip())
```
